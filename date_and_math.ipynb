## Python Maths

```python
# list1=[10,5,15,20]
# x= min(list1)
# print(x)
x=min(10,5,15,20)
y=max(10,5,20,15)
print("min:",x)
print("max:",y)
```

```python
#absolute value
x= abs(-7)
y= abs(-7.25)
print(x)
print(y) # minus convert to positive
```

```python
#power
x= pow(4,3) # 4*4=16*4=64 (4*4*4)
print(x)
y= pow(10,5) # 10 *10 * 10 *10 *10 = 100000
print(y)
```

```python
import math
x= math.sqrt(64) # square of 8 in 64  (8*8=64)
y= math.sqrt(16) # square of 4 in 16  (4*4=16)
print(x)
print(y)
```

```python
import math
x1= math.ceil(7.3)   # ceil upper value batave
print(x1) 
x1= math.ceil(7.8)   
print(x1) 

y1= math.floor(7.3)   # ceil upper value batave
print(y1) 
y1= math.floor(7.8)   
print(y1) 
```
