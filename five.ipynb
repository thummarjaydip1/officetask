## Python String Formatting

```python
txt= "this is string example 49"
print(txt)
```

```python
x=49
txt = "this is string example {x}"
print(txt)
txt = f"this is string example {x}"
print(txt)
```

```python
division= 55/2
txt = f"this is division {division:.2f}"
print(txt)
```

```python
txt= f"this is price of dollar {95:.2f}"
print(txt)
```

```python
txt= f"this is price of 10 dollar {10*95:.2f}"
print(txt)
```

```python
price = 50
tax = 0.25
txt= f"this is string {price+ (price * tax):.2f}"
print(txt)
```

```python
price = 49
txt= f"it is a very {'expensive' if (price>=50) else 'cheap'}"
print(txt)
```

```python
fruit = "apple"
txt = f"i love {fruit.upper()}"
print(txt)
```

```python
def myfun(rupees):
    return rupees * 100 ,"paisa",rupees ,"rupees"
txt = f"rupees convert to paisa {myfun(100)}"
print(txt)
```
