## Python JSON

###### json data read own file ==> loads() method
###### json data read another .json file => load() method
###### json data write own file ==> dumps() method
###### json data write another .json file ==> dump() method

```python
import json
x = '{"name":"jaydip", "age":20, "city":"amreli"}'
y = json.loads(x)
print(y["name"])
print(y["city"])
print(y)
```

```python
#dumps method dictonary convert to json(string)
import json
x= {
    "name" : "jaydip",
    "city" : "rajkot",
    "phone no": 9714032218,
    "email-id" : "thummarjaydip1@gmail.com"
} 
print(type(x))
y= json.dumps(x)
print(y)
# print(y['city']) #generate a error
print(type(y))
```

```python
import json  #all data type convert to json(string)

print(json.dumps(["apple","banana"]))
print(json.dumps(("apple","banana")))
print(json.dumps({"name":"jaydip","age":23}))
print(json.dumps(12345))
print(json.dumps(12.34))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
import json
x ={
    "name" : "mayur",
    "age" : 33,
    "merried" : True,
    "divorces" : False,
    "children": ("manav", "manshi"),
    "pet" : None,
    "car" :[
        {"model":"zen", "year":2022},
        {"model":"aulto", "year":2020},
        {"model":"honda", "year":2018}
    ]
}
print(json.dumps(x, indent=3,separators=("."," = "),sort_keys=False))
#indent = 3 means 3 space to left side
#separators = ("."," = ") means key and value separate by " = " and item separate by "."
#sort_keys = False means not sorted by key and True means sorted by key
```

```python
# write json file in one.json file 
import json
x= {
    "name": "jaydip",
    "age": 20,
    "city": "amreli",
    "phone no": 9714032218,
    "email-id": "thummarjaydip26@gmail.com"
}
with open("one.json","w") as file:
    json.dump(x, file, indent=3)
print("data write successfully")    
```

## Python RegEX

######     . all character 
######     r no means raw string
######     \d no means digit (0-9)
######     \D no means not digit
######     \s no means space
######     ^ means start 
######     $ means end
######     + means one or more
######     * means zero or more
######     ? means zero or one

```python
import re
txt = "the rain in Spain"
x=re.search("^the.*in$", txt)
print(x)
y = re.search("different", "I have five different cats")
print(y)
```

## Python Exception Handling 

###### try ==> error generate and manage
###### except ==> error handle
###### else ==> error not generate to else execute other wise not execute 
###### finally ==> always execute

```python
try:
    print(x123)
except:
    print("An exception occurred")
```

```python
try:
    print(f"hello {1}")   
except:
    print("an exception generate") 
else:
    print("not generate exception") #not generate to exception this block are executed
finally:
    print("this is finally block") #always printed
```

```python
try:
    ab= "hello"
    # print(int(abc)) #name error
    print(ab) #value error
except NameError:
    print("variable are not defined variable name error")
except ValueError:
    print("value not defined to int value error")
except:
    print("somthing went wrong")
```

```python
x = -1
if x<0:
    raise Exception("number is below to zero")
else:
    raise Exception("number is above to zero")
```

```python
x="hello"
if not type(x) is int:
    raise TypeError("only integers are allowed")
```

```python
try:
    a= int(input("enter first number: "))
    b= int(input("enter second number: "))
    x= a/b
except ZeroDivisionError:
    print("divided by zero is not allow")
except:
    print("another exception generate")
else:
    print(x)
```
