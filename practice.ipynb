```python
def myfun(row,col):
    for i in range(row):
        for j in range(col):
            if i==0 or i==row-1 or j==0 or j==col-1:
                print("*", end=" ")
            else:
                print(" ",end=" ")
        print()
myfun(5,5)
```

```python
def myfun(*numbers):
    if numbers == 0:
        return None
    max_no = numbers[0]
    for num in numbers:
        if num > max_no:
            max_no = num
    return max_no
print(myfun(10,20,30,40))
```

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))
```

```python
def fibonacii(n):
    if n <= 1:
        return n
    else:
        return fibonacii(n-1) + fibonacii(n-2)
print(fibonacii(7))
```

```python
def sum_list(number):
    if len(number) == 0:
        return 0
    else:
        return number[0] + sum_list(number[1:])
num = [10,20,30]
print(sum_list(num))
```

```python
def find_max(number):
    if len(number) == 1:
        return number[0]
    else:
        max_no = find_max(number[1:])
        return number[0] if number[0] > max_no else max_no
num= [10,5,65,34,2,87,8,6,34]
print(find_max(num))
```

```python
def myfun(a):
    try:
        a=int(a)
    except ValueError:
        print("value error generated")
    except:
        print("error generated")
    else:
        if a%4==0:
            leap= list(filter(lambda a:a%4==0, [a]))
            print("leap year",leap)
        else:
            not_leap= list(filter(lambda a:a%4!=0, [a]))
            print("not leap",not_leap)
    finally:
        print("leap year program")
x= input("enter year to check leap year or not:")
myfun(x)
```
