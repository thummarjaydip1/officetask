# Power Calculator

# Simple Login Check
#### Set username/password
#### Ask user input and check if correct

```python
username="jaydip"
password="1234"
user=input("enter username:")
passw=input("enter password:")
if username==user and password==passw:
    print("username and password matched")
elif username!=user and password==passw:
    print("username different or password matched")
elif username==user and password!=passw:
    print("username matched or password different")
else:
    print("username and password are not matched")
```

# Mini Grade System
#### Input marks
#### Print grade:
- 90+ = A
- 70–89 = B
- 50–69 = C
- below 50 = Fail

```python
a=int(input("enter mark:"))
if a>=90:
    print("Grade: A")
elif a<=89 and a>=70:
    print("grade B")
elif a<=69 and a>=50:
    print("Grade C")
else:
    print("Fail")
```
