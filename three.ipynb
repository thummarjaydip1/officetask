##### Python Decorators fun

```python
def changecase(fun):
    def myinner():
        return fun().upper()
    return myinner
@changecase
def myfun():
    return "hello world"
print(myfun())
```

```python
def changecaseupper(fun):
    def myinner():
        return fun().upper()
    return myinner
    
def changecaselower(fun):
    def inner():
        return fun().lower()
    return inner

@changecaseupper
def myfun1():
    return "hello"

@changecaselower
def myfun2():
    return "WORLD"
print(myfun1())
print(myfun2())
```

```python
def changecase(fun):
    def myinner(abc):
        return fun(abc).upper()
    return myinner

@changecase
def myfun(name):
    return "hello " + name
print(myfun("jaydip"))
```

```python
def changecase(fun):
    def inner(*args,**kwargs):
        return fun(*args,**kwargs).upper()
    return inner
@changecase
def myfun(name):
    return "hello " + name
# print(myfun(name="jay"))
print(myfun("jay"))
```

```python
def changecase(n):
    def changecase(fun):
        def inner():
            if n==1:
                a= fun().lower()
            else:
                a= fun().upper()
            return a
        return inner
    return changecase
@changecase(1)
def myfun():
    return "Hello World"
print(myfun())
```

```python
def changecase(fun):
    def inner():
        return fun().upper()
    return inner

def addgreeting(fun):
    def inner():
        return "hello " + fun() + " have good day"
    return inner

@changecase
@addgreeting
def myfun():
    return "Jaydip"
print(myfun())
```

##### Python Recursion

```python
def countdown(n):
    if n<=0:
        print("Done")
    else:
        print(n)
        countdown(n-1)
countdown(5)
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(6))
```

```python
def sum_list(numbers):
    if len(numbers) == 0:
        return 0
    else:
        return numbers[0] + sum_list(numbers[1:])
num = [1,2,3,4,5]
print(sum_list(num))
```

```python
def find_max(number):
    if len(number) == 1:
        return number[0]
    else:
        max_rest = find_max(number[1:])
        return number[0] if number[0] > max_rest else max_rest
num= [23,45,89,23,65,99,33]
print(find_max(num))
```

```python
def myfun():
    yield 1
    yield 2
    yield 3
for value in myfun():
    print(value)
```

```python
def count_up(n):
    count=1
    while count<=n:
        yield count
        count+=1
for num in count_up(4):
    print(num)
```

```python
def large_seq(n):
    for i in range(n):
        yield i
gen=large_seq(10000000)
print(next(gen))
print(next(gen))
print(next(gen))
```
