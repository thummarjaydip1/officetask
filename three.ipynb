## Python Range

```python
x= range(10)
print(x)
print(list(x))
```

```python
# range(start_number,stop_number,increment/decrement_number)
x= range(2,20,2)
print(x)
print(list(x))
```

```python
for i in range(10):
    print(i, end=" ")
```

```python
print(list(range(11)))
print(list(range(5,16)))
print(list(range(10,101,10)))
```

```python
a=range(10)
print(a[4])
print(a[:4])
print(list(a[:4]))
```

```python
a= range(0,20,2)
print(list(a))
print(6 in a)
print(7 in a)
print(len(a))
```

## Python Array

```python
car= ['ford','shift','bwm','aulto','fortuner','volvo']
print(car)
```

```python
car= ['ford','shift','bwm','aulto','fortuner','volvo']
car[3] = 'toyota'
print(car)
print("length of array:",len(car))
```

```python
car= ['ford','shift','bwm','aulto','fortuner','volvo']
car.append('honda')
print(car)
```

```python
car= ['ford','shift','bmw','aulto','fortuner','volvo','honda']
new = ['shift','aulto','zen','honda']
car.pop(3) # auto remove index 3
car.pop() # last item removed
car.remove("shift") # shift removed
print(car.count("volvo")) #volvo ne count kare 1 time che
print(car)
car.extend(new)
print("extended array:",car)
car.sort()            #reverse=True  reverse sorting thay array
print("sorted array:",car) 
car.reverse()
print("reverse sorted:",car)
print(car.index("shift"))
```

```python
def myfun(e):
    return e['year']
cars=[
    {'car':'ford','year':2007},
    {'car':'toyota','year':2000},
    {'car':'bmw','year':2010},
    {'car':'volvo','year':2003}
]
cars.sort(key=myfun)
print(cars)
```

```python
def myfun(e):
    return len(e)
cars= ['toyota','bmw','volvo','fortuner','shift']
cars.sort(key=myfun)
print(cars) #length wise soted array
cars.sort(reverse=True, key=myfun)
print(cars) #length wise soted array
```

```python
class Myclass:
    def __iter__(self):
        self.a=1
        return self
    def __next__(self):
        x = self.a
        self.a += 1
        return x
myobj = Myclass()
myiter = iter(myobj)
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```

```python
class Myclass:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myobj = Myclass()
myiter = iter(myobj)

for x in myiter:
  print(x)
```
