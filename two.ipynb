# Python Match

```python
day= 4
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
```

```python
num = 9
match num:
    case 1:
        print("Today is one")
    case 2:
        print("Today is two")
    case 3:
        print("Today is three")
    case 4:
        print("Today is four")
    case 5:
        print("Today is five")
    case 6:
        print("Today is six")
    case 7:
        print("Today is seven")
    case _:
        print("another no")
```
